```python
# Define the columns for outlier detection
columns_to_check = ['gross_dollars', 'return_dollars', 'net_after_return_dollars',
                    'trade_dollars', 'net_after_trade_dollars', 'gross_units', 'net_units']

def impute_outliers_with_median(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)  # First quartile (25%)
        Q3 = df[column].quantile(0.75)  # Third quartile (75%)
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Calculate the median of the column
        median_value = df[column].median()

        # Replace outliers with the median
        df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound),
                              median_value, df[column])
    return df

# Apply the function to impute outliers
df = impute_outliers_with_median(df, columns_to_check)

# Check the result
df.describe()
```

```python
# Drop unnecessary columns (e.g., unnamed index columns if present)
df = df.drop(columns=['unnamed:_0'], errors='ignore')
```

```python
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Save the encoder for inverse transformation if needed
```

```python
# ----> Extract numerical features from datetime columns <----
df['fulldate_year'] = df['fulldate'].dt.year
df['fulldate_month'] = df['fulldate'].dt.month
df['fulldate_day'] = df['fulldate'].dt.day

df['formatteddate_year'] = df['formatteddate'].dt.year
df['formatteddate_month'] = df['formatteddate'].dt.month
df['formatteddate_day'] = df['formatteddate'].dt.day
```

```python
# Drop original datetime columns
df = df.drop(columns=['fulldate', 'formatteddate'])
```

```python
# Split data into features (X) and target variables (y)
X = df.drop(columns=['gross_dollars', 'net_units'], errors='ignore')  # Features
```

```python
# Regression Task: Predicting gross_dollars
y_regression = df['gross_dollars']
```

```python
# Identify categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns
```

```python
# Create a ColumnTransformer to apply OneHotEncoding to categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', X.select_dtypes(exclude=['object']).columns.tolist()),  # Passthrough numerical features
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_cols.tolist()), # One-hot encode categorical features
    ])
```

```python
# Apply the preprocessing to your data
X_processed = preprocessor.fit_transform(X)
```

```python
# Now split the preprocessed data
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_processed, y_regression, test_size=0.2, random_state=42)
```

```python
X_train_reg
```

```python
# Initialize Decision Tree models
regressor = DecisionTreeRegressor(max_depth=5, random_state=42)
```

```python
# Train models
regressor.fit(X_train_reg, y_train_reg)
```

```python
# Predictions
y_pred_reg = regressor.predict(X_test_reg)
```

```python
# Evaluate Regression Model
mse = mean_squared_error(y_test_reg, y_pred_reg)
print(f"Regression Model MSE: {mse}")
```

```python
r2 = r2_score(y_test_reg, y_pred_reg)
print(f"Regression Model R-squared: {r2}")
```

```python
# Display Results
results = {
    "Regression": {
        "Mean Squared Error": mse,
        "RÂ² Score": r2
    }
}  # Make sure this closing brace is aligned with 'results ='
results
```

```python
# Classification Task: Categorizing sales into 'high' and 'low' based on median
sales_median = df['gross_dollars'].median()
df['sales_category'] = df['gross_dollars'].apply(lambda x: 'high' if x > sales_median else 'low')
```

```python
# Convert categorical target variable to numerical format
y_classification = df['sales_category'].astype('category').cat.codes
```

```python
# Split the preprocessed data
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_processed, y_classification, test_size=0.2, random_state=42)
```

```python
# Initialize Decision Tree Classifier
classifier = DecisionTreeClassifier(max_depth=5, random_state=42)
```

```python
# Train model
classifier.fit(X_train_cls, y_train_cls)
```

```python
# Predictions
y_pred_cls = classifier.predict(X_test_cls)
```

```python
# Initialize Decision Tree Classifier with aggressive constraints to prevent overfitting
classifier = DecisionTreeClassifier(
    max_depth=3,               # Limit depth further
    min_samples_split=100,     # Require at least 100 samples to split
    min_samples_leaf=50,       # Each leaf must have at least 50 samples
    max_features="sqrt",       # Consider only a subset of features at each split
    ccp_alpha=0.05,            # Increase pruning strength
    random_state=42
)

# Train the classifier
classifier.fit(X_train_cls, y_train_cls)

# Predictions
y_pred_cls = classifier.predict(X_test_cls)

# Evaluate Classification Model
accuracy = accuracy_score(y_test_cls, y_pred_cls)
print(f"Classification Model Accuracy: {accuracy}")

classification_rep = classification_report(y_test_cls, y_pred_cls, target_names=['Low Sales', 'High Sales'])
print("Classification Report:\n", classification_rep)
```

```python


et_classifier = ExtraTreesClassifier(
    n_estimators=50,           # Reduce number of trees
    max_depth=3,               # Lower depth to prevent memorization
    min_samples_split=200,     # Require more samples to split
    min_samples_leaf=100,      # Force larger leaf nodes
    max_features="sqrt",       # Limit the number of features at each split
    class_weight="balanced",   # Handle class imbalance
    random_state=42
)

# Train the model
et_classifier.fit(X_train_cls, y_train_cls)

# Predictions
y_pred_et = et_classifier.predict(X_test_cls)

# Evaluate Model
accuracy_et = accuracy_score(y_test_cls, y_pred_et)
print(f"Extra Trees Model Accuracy: {accuracy_et}")

classification_rep_et = classification_report(y_test_cls, y_pred_et, target_names=['Low Sales', 'High Sales'])
print("Classification Report (Extra Trees):\n", classification_rep_et)
```
